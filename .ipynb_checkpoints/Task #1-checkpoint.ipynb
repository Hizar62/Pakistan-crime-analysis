```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('crime.csv')
```

```python
data=data.drop(['_id','Railways','Pakistan'],axis=1)
```

```python
data.describe()
```

```python
data.hist()
```

```python
x = data['Year']
y = data['Punjab']
plt.bar(x,y, label = 'Punjab', width =0.4, color='r')
plt.title('Crime Rate in Punjab in different Years')
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
x = data['Offence']
y = data['Punjab']
plt.bar(x,y, label = 'Punjab', width =0.4, color='r')
plt.title('Offence rate in Punjab')
plt.xlabel('Offense- Crime Name')
plt.ylabel('Frequency')
plt.xticks(rotation=80)
plt.legend()
plt.show()
```
